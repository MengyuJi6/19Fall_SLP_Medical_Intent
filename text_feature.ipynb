Option 1 : Use unique text to extract features

```python
import numpy as np
rows_list = []
for index in range(len(final)):
    #file_name = final.iloc[index]["file_name"]
    phrase = final.iloc[index]["phrase"]
    prompt = final.iloc[index]["prompt"]
    
    tokens = nltk.word_tokenize(phrase)
    tagged = nltk.pos_tag(tokens)
    skeleton_dict_copy = skeleton_dict.copy()
    for i in range(len(tagged)):
        skeleton_dict_copy[tagged[i][1]] = skeleton_dict_copy[tagged[i][1]]+1
    #skeleton_dict_copy["file_name"] = file_name
    skeleton_dict_copy["phrase"] = phrase
    skeleton_dict_copy["prompt"] = prompt
    rows_list.append(skeleton_dict_copy)
df = pd.DataFrame(rows_list)
```

Option 2: use all text (including duplicate text) to extract features

```python
import numpy as np
rows_list = []
from tqdm import tqdm
t = tqdm(total=len(csv))
for index in range(len(csv)):
    file_name = csv.iloc[index]["file_name"]
    phrase = csv.iloc[index]["phrase"]
    prompt = csv.iloc[index]["prompt"]
    
    tokens = nltk.word_tokenize(phrase)
    tagged = nltk.pos_tag(tokens)
    skeleton_dict_copy = skeleton_dict.copy()
    for i in range(len(tagged)):
        skeleton_dict_copy[tagged[i][1]] = skeleton_dict_copy[tagged[i][1]]+1
    skeleton_dict_copy["file_name"] = file_name
    skeleton_dict_copy["phrase"] = phrase
    skeleton_dict_copy["prompt"] = prompt
    rows_list.append(skeleton_dict_copy)
    t.update(1)
df = pd.DataFrame(rows_list)
```

```python
#train_textX=train[keyList]
train_textX=train[features]
train_texty=train["prompt"]
train_textX = np.array(train_textX.values.tolist())
train_texty = np.array(train_texty.tolist())
```

```python
from sklearn import multiclass
from sklearn import tree
from sklearn.svm import LinearSVC
```

```python
classifier = multiclass.OneVsRestClassifier(tree.DecisionTreeClassifier())
classifier.fit(train_textX,train_texty)
```

```python
#test_textX=test[keyList]
test_textX=test[features]
test_texty=test["prompt"]
test_textX = np.array(test_textX.values.tolist())
test_texty = np.array(test_texty.tolist())
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
predict_texty = classifier.predict(test_textX)
accuracy=accuracy_score(test_texty, predict_texty)
f1=f1_score(test_texty, predict_texty, average="weighted")
cm=confusion_matrix(test_texty, predict_texty)
print(accuracy)
print(f1)
print(cm)
```
